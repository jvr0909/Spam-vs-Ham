```python
data = pd.read_csv("../input/mailing/spamham.csv", thousands=',')
data.head(3)

#Note: When importing the data from the zip file into kaggle it said this was a similar dataset so I used the one kaggle refrenced.
```

```python
ytrain,ytest,xtrain,xtest = train_test_split(data["Category"],data["Message"], test_size = 0.2 , random_state = 42)
xtrain.head(3)
```

```python
vect = CountVectorizer(stop_words='english') # text into a vector on the basis of the count. https://www.geeksforgeeks.org/using-countvectorizer-to-extracting-features-from-text/
vect.fit(xtrain)
```

```python
# print(vect.get_feature_names())
# print(vect.get_feature_names())
```

```python
xtrain2 = vect.transform(xtrain) #build matrix from training data
xtest2 = vect.transform(xtest) # ignores tokens it hasn't seen before
type(xtest2)
```

```python
from sklearn.metrics import confusion_matrix
prediction = dict()
final = MultinomialNB()
final.fit(xtrain2,ytrain)
prediction["naive_bayes"] = final.predict(xtest2)
confusion_matrix(ytest, prediction['naive_bayes'])
```
